
### Task One: Creating data from an equation

It is important to be able to directly translate a real equation into a plot. Your first task actually is pure numpy, then we will explore how to plot it out with Matplotlib. The [world famous equation](https://en.wikipedia.org/wiki/Mass%E2%80%93energy_equivalence) from Einstein:

$$E=mc^2$$



```python
# CODE HERE
import numpy as np
import matplotlib.pyplot as plt
m = np.linspace(0, 10, 11)
m
```

```python
# CODE HERE
#e = mc^2
c = 300000000
e = m * c**2
e
```

### Part Two: Plotting E=mc^2


```python
# CODE HERE
plt.plot(m,e, color='red',lw=5, marker='x')
plt.title("E = MC^2")
plt.xlabel("Mass in grams")
plt.ylabel("Energy in joules")
plt.xlim(0, 10)
plt.show()
```

### Part Three (BONUS)

**Plot on a logarthimic scale on the y axis**

```python
# CODE HERE
plt.plot(m,e, color = 'red', lw=5)
plt.xlabel('Mass in Grams')
plt.ylabel('Energy in joules')
plt.yscale('log')
plt.title('E = MC^2')
plt.grid(which='both',axis='y')
plt.show()
```



## Task Two: Creating plots from data points

In finance, the yield curve is a curve showing several yields to maturity or interest rates across different contract lengths (2 month, 2 year, 20 year, etc. ...) for a similar debt contract. The curve shows the relation between the (level of the) interest rate (or cost of borrowing) and the time to maturity, known as the "term", of the debt for a given borrower in a given currency.

The U.S. dollar interest rates paid on U.S. Treasury securities for various maturities are closely watched by many traders, and are commonly plotted on a graph such as the one on the right, which is informally called "the yield curve".

## Part One: Yield Curve Data


```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

**TASK: How to plot both curves on the same Figure. Add a legend to show which curve corresponds to a certain year.**

```python
# CODE HERE
plt.plot(labels,july16_2007 , label = 'july16_2007')
plt.plot(labels,july16_2020 , label = 'july16_2020')
plt.legend()
```

**The legend in the plot above looks a little strange in the middle of the curves. While it is not blocking anything, it would be nicer if it were *outside* the plot. How to move the legend outside the main Figure plot.**

```python
# CODE HERE# CODE HERE
plt.plot(labels,july16_2007 , label = 'july16_2007')
plt.plot(labels,july16_2020 , label = 'july16_2020')
plt.legend(loc =(1.1,0.5))
```

**While the plot above clearly shows how rates fell from 2007 to 2020, putting these on the same plot makes it difficult to discern the rate differences within the same year. Using .suplots() to create the plot figure below, which shows each year's yield curve.**

```python
# CODE HERE
fig, ax = plt.subplots(nrows=2, figsize = (12,8))
ax[0].plot(labels,july16_2007)
ax[0].set_title('July 16th, 2007')

ax[1].plot(labels,july16_2020)
ax[1].set_title('July 16th, 2020')
plt.show()
```

**Create the plot below that uses twin axes.**

```python
# CODE HERE
fig, ax = plt.subplots()
ax.plot(labels,july16_2007, color = 'b')
ax.set_ylabel('2007', fontsize=14, color = 'b')
ax.spines['left'].set_color('blue')
ax.spines['left'].set_linewidth(4)
for label in ax.get_yticklabels():
    label.set_color('b')
plt.yticks(fontsize=16)

ax2 = ax.twinx()
ax2.plot(labels,july16_2020, color = 'r')
ax2.set_ylabel('2020', fontsize=14, color='r')
ax2.spines['right'].set_color('r')
ax2.spines['right'].set_linewidth(4)
for label in ax2.get_yticklabels():
    label.set_color('r')
plt.yticks(fontsize=16)

plt.title('July 16th, curves')
plt.show()
```
